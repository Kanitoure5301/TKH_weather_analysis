```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isnull()
```

```python
# TODO: drop your null values
cleaned_coffee=coffee.dropna()
```

```python
# TODO: Select only Minas Gerais

mg_coffee= cleaned_coffee[cleaned_coffee['subdivision'] == 'Minas Gerais']
```

```python
# TODO: BEGIN YOUR EDA!

# Plotting Bearing trees across the years 
plt.plot(mg_coffee['year'],mg_coffee['bearing_trees'],marker='o', linestyle='-')
plt.title('Bearing Trees Change Throughout Time')
plt.xlabel('Years')
plt.ylabel('Bearing Trees')
plt.show()

# Plotting Bearing Hectares across the years

plt.plot(mg_coffee['year'],mg_coffee['bearing_trees'],marker='o', linestyle='-')
plt.title('Bearing Hectares Change Throught Time')
plt.xlabel('Years')
plt.ylabel('Bearing Hectares')
plt.show()

# Plotting Count of Coffee Change 

plt.plot(mg_coffee['year'],mg_coffee['60kgs_bag'],marker='o', linestyle='-')
plt.title('60 Kilogram Bags of Coffee Over Time (Minas Gerais)')
plt.xlabel('Years')
plt.ylabel('60 Kilogram Bags (Millions)')
plt.show()

# Plotting Relationship Between Bearing Trees and 60kgs bags

Correlation1= mg_coffee['bearing_trees'].corr(mg_coffee['60kgs_bag'])
plt.scatter(mg_coffee['bearing_trees'], mg_coffee['60kgs_bag'])
plt.title('Bearing Trees Vs 60 kilogram bags')
plt.xlabel('Bearing Trees (Millions)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
print("Pearson correlations=", Correlation1)

# Graphing he relationship between bearing_hectares and 60kgs_bag


Correlation2= mg_coffee['bearing_hectares'].corr(mg_coffee['60kgs_bag'])
plt.scatter(mg_coffee['bearing_hectares'], mg_coffee['60kgs_bag'])
plt.title('Bearing Hectares Vs 60 kilogram bags')
plt.xlabel('Bearing Hectares (Thousand Hectares)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
print("Pearson correlations=", Correlation2)
```
