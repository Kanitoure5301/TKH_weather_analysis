```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import coffee data
coffee = pd.read_csv("../data/crop/coffee_output.csv")

# import weather data
weather=pd.read_csv("../data/weather/weather_data.csv")
```

```python
# TODO: Merge these two dataframes together

ow_merged = pd.merge(coffee, weather, on=['year', 'subdivision'])
```

```python
# TODO: View the first 5 rows of your data

ow_merged.head()
```

```python
# TODO: Check for null values 

ow_merged.isnull()
```

```python
# TODO: Drop rows that contain null values

cleaned_ow= ow_merged.dropna()

cleaned_ow.head()
```

```python
# TODO: BEGIN YOUR EDA!

# Plotting relationship between temp_avg and 60kgs_bag?

plt.scatter(cleaned_ow['temp_avg'], cleaned_ow['60kgs_bag'])
plt.title('Average Temp Vs 60 kilogram bags')
plt.xlabel('Temp (C)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
Correlation1= cleaned_ow['temp_avg'].corr(cleaned_ow['60kgs_bag'])
print("Pearson correlation=", Correlation1)

# Plotting relationship between minimum humidity and 60kgs_bag
plt.scatter(cleaned_ow['hum_min'], cleaned_ow['60kgs_bag'])
plt.title('Minimum Humidity Vs 60 kilogram bags')
plt.xlabel('Minimum Humidity (%)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
Correlation2= cleaned_ow['hum_min'].corr(cleaned_ow['60kgs_bag'])
print("Pearson correlation=", Correlation2)

# Plotting relationship between maximum rain and 60kg bags
plt.scatter(cleaned_ow['rain_max'], cleaned_ow['60kgs_bag'])
plt.title('Maximum Rain Vs 60 kilogram bags')
plt.xlabel('Maximum Rain (mm)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
Correlation3= cleaned_ow['rain_max'].corr(cleaned_ow['60kgs_bag'])
print("Pearson correlation=", Correlation3)

# Plotting relationship between average wind speed and 60kg bags
plt.scatter(cleaned_ow['wind_avg'], cleaned_ow['60kgs_bag'])
plt.title('Wind Average Vs 60 kilogram bags')
plt.xlabel('Wind Average (Meters Per Second)')
plt.ylabel('60 Kilograms Bags (Millions)')
plt.show()
Correlation3= cleaned_ow['wind_avg'].corr(cleaned_ow['60kgs_bag'])
print("Pearson correlation=", Correlation3)

# Weather Analysis vs Bearing Trees
plt.scatter(cleaned_ow['temp_avg'], cleaned_ow['bearing_trees'])
plt.title('Average Temperature Vs Bearing Trees')
plt.xlabel('Temp (C)')
plt.ylabel('Bearing Trees (Millions)')
plt.show()
Correlation4= cleaned_ow['temp_avg'].corr(cleaned_ow['bearing_trees'])
print("Pearson correlation=", Correlation4)

plt.scatter(cleaned_ow['hum_min'], cleaned_ow['bearing_trees'])
plt.title('Minimum Humidity Vs Bearing Trees')
plt.xlabel('Minimum Humidity (%)')
plt.ylabel('Bearing Trees (Millions)')
plt.show()
Correlation5= cleaned_ow['hum_min'].corr(cleaned_ow['bearing_trees'])
print("Pearson correlation=", Correlation5)

plt.scatter(cleaned_ow['rain_max'], cleaned_ow['bearing_trees'])
plt.title('Maximum Rain Vs Bearing Trees')
plt.xlabel('Maximum Rain (mm)')
plt.ylabel('Bearing Trees (Millions)')
plt.show()
Correlation6= cleaned_ow['rain_max'].corr(cleaned_ow['bearing_trees'])
print("Pearson correlation=", Correlation6)


plt.scatter(cleaned_ow['wind_avg'], cleaned_ow['bearing_trees'])
plt.title('Wind Average Vs Bearing Trees')
plt.xlabel('Wind Average (Meters Per Second)')
plt.ylabel('Bearing Trees (Millions)')
plt.show()
Correlation7= cleaned_ow['wind_avg'].corr(cleaned_ow['bearing_trees'])
print("Pearson correlation=", Correlation7)
```
