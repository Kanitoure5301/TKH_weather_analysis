```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1,weather2],ignore_index=True)
weather.to_csv("../data/weather/weather_data.csv", index=False)
```

```python
# TODO: view first 5 rows of your data
weather.head()
```

```python
# TODO: describe your numeric columns
weather.describe()
```

```python
# TODO: display the null values in your dataframe
weather.isnull()
```

```python
# TODO: BEGIN YOUR EDA!

#Ploting the average temperature by year

plt.plot(weather['year'],weather['temp_avg'],marker='o', linestyle='-')
plt.title('The Average Temperature Over The Years')
plt.xlabel('Year')
plt.ylabel('Average Temperature (C)')
plt.grid(True)
plt.show()


#Ploting Humidity Trends using scatterplot

plt.plot(weather['year'],weather['hum_min'],marker='o', linestyle='-', color = 'Green')
plt.title('Minimum Humidity Over the Years (Minas Gerais)')
plt.xlabel('Year')
plt.ylabel('Minimun Humidity (%)')
plt.grid(True)
plt.show()

#Line Plot for Wind Speeds by Year

plt.plot(weather['year'], weather['wind_avg'], marker='o', linestyle='-', color='red')
plt.title('Average Wind Speeds By Year')
plt.xlabel('Year')
plt.ylabel('Wind Speed')
plt.grid(True)
plt.show()

#Line Chart for Highest RainFall

plt.plot(weather['year'], weather['rain_max'], marker='o', linestyle='-', color='orange')
plt.title('Maximum Rainfall Over the Years')
plt.xlabel('Year')
plt.ylabel('Max Rainfall')
plt.grid(True)
plt.show()

#Graphing Relationship between max rain and max humidity using Pearson's Correlation

correlation1= weather['rain_max'].corr(weather['hum_max'])
sns.lmplot(x='rain_max', y='hum_max', data=weather)
plt.title('Max Rainfall vs Max Humidity')
plt.xlabel('Max Rainfall')
plt.ylabel('Max Humidity')
plt.show()
print("Pearson correlations=", correlation1)


#Graphing Relationship between Avg Wing vs Avg Temp
correlation2= weather['wind_avg'].corr(weather['temp_avg'])
plt.scatter(weather['wind_avg'], weather['temp_avg'])
plt.title('Average Wind Speed vs Average Temp')
plt.xlabel('Avg Wind Speed')
plt.ylabel('Avg Temp (C)')
plt.show()
print("Pearson correlations=", correlation2)
```
